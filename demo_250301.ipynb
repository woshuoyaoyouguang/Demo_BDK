```python
import pandas as pd
import numpy as np
from scipy import stats
from sklearn.metrics import mean_squared_error
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split, KFold
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python

# 1. Pearson 相关系数
def calculate_pearson_corr(y_true, y_pred, weight=None):
    """计算 Pearson 相关系数（可选加权）"""
    def weighted_avg(x, w):
        return np.average(x, weights=w) if w is not None else np.mean(x)
    
    def weighted_cov(x, y, w):
        return np.sum(w * (x - weighted_avg(x, w)) * (y - weighted_avg(y, w))) / np.sum(w) if w is not None else np.cov(x, y)[0, 1]
    
    return weighted_cov(y_true, y_pred, weight) / (np.sqrt(weighted_cov(y_true, y_true, weight)) * np.sqrt(weighted_cov(y_pred, y_pred, weight)))

# 2. Spearman 相关系数
def calculate_spearman_corr(y_true, y_pred):
    """计算 Spearman 相关系数"""
    return stats.spearmanr(y_true, y_pred)[0]

# # 3. Gini 系数
# def calculate_gini(y_true, y_pred):
#     """计算 Gini 系数"""
#     sorted_indices = np.argsort(y_pred)
#     sorted_true = y_true.iloc[sorted_indices]
    
#     cum_true = np.cumsum(sorted_true) / np.sum(sorted_true)
#     rand = np.linspace(0, 1, len(y_true))

#     gini = 2 * np.mean(cum_true - rand)
#     return gini

# def calculate_mod_gini(y_true, y_pred, sample_weight=None):
#     """正确的 Modified Gini 计算方式"""
#     y_true = np.array(y_true)
#     y_pred = np.array(y_pred)
    
#     # 排序索引
#     sorted_indices = np.argsort(y_pred)
#     sorted_true = y_true[sorted_indices]

#     # 计算累计加权事件数
#     if sample_weight is None:
#         sample_weight = np.ones_like(y_true)  # 默认权重为 1
    
#     sorted_weight = sample_weight[sorted_indices]
#     weighted_event = sorted_true * sorted_weight
#     cum_event = np.cumsum(weighted_event)
#     total_event = np.sum(weighted_event)

#     # 计算 Lorenz 曲线 (真实事件累积分布)
#     cum_true = cum_event / total_event

#     # 计算随机分布 (rand)
#     rand = np.linspace(0, 1, len(y_true))

#     # 计算理论最优 Lorenz 曲线
#     best = np.minimum(np.cumsum(sorted_weight), np.asarray([total_event] * len(y_true))) / total_event

#     # 避免分母为 0
#     denom = np.mean(best - rand)
#     if denom == 0:
#         return np.nan  # 避免 0 除错误

#     return np.mean(cum_true - rand) / denom

def gini(actual, pred):
    """计算标准 Gini 系数"""
    assert (len(actual) == len(pred))
    all = np.asarray(np.c_[actual, pred, np.arange(len(actual))], dtype=np.float64)
    all = all[np.lexsort((all[:, 2], -1 * all[:, 1]))]  # 先按 pred 降序排序，再按索引稳定排序
    totalLosses = all[:, 0].sum()
    giniSum = all[:, 0].cumsum().sum() / totalLosses

    giniSum -= (len(actual) + 1) / 2.  # 校正
    return giniSum / len(actual)

def gini_normalized(actual, pred):
    """计算标准化 Gini 系数 (Modified Gini)"""
    return gini(actual, pred) / gini(actual, actual)


# 5. KS 统计量
def calculate_ks(y_true, y_pred):
    """计算 KS 统计量"""
    data = np.vstack([y_pred, y_true]).T
    sorted_data = data[np.argsort(data[:, 0])]

    cum_positive_rate = np.cumsum(sorted_data[:, 1]) / np.sum(sorted_data[:, 1])
    cum_negative_rate = np.cumsum(1 - sorted_data[:, 1]) / np.sum(1 - sorted_data[:, 1])
    
    ks = max(abs(cum_positive_rate - cum_negative_rate))
    return ks

# 6. RMSE
def calculate_rmse(y_true, y_pred):
    """计算 RMSE"""
    return np.sqrt(mean_squared_error(y_true, y_pred))

# 7. Hosmer-Lemeshow 统计量
def calculate_hosmer_lemeshow(y_true, y_pred, n_groups=10):
    """计算 Hosmer-Lemeshow 统计量"""
    data = pd.DataFrame({'y_true': y_true, 'y_pred': y_pred})
    data['deciles'] = pd.qcut(data['y_pred'], n_groups, duplicates='drop')

    grouped = data.groupby('deciles')
    obs = grouped['y_true'].sum()
    total = grouped['y_true'].count()
    expected = grouped['y_pred'].sum()

    hl_stat = (((obs - expected) ** 2) / expected + ((total - obs - (total - expected)) ** 2) / (total - expected)).sum()
    return hl_stat

# 8. 准确率指数
def calculate_accuracy_index(y_true, y_pred):
    """计算准确率指数"""
    return np.mean(abs(y_true - y_pred))

# 9. Lorenz 曲线（用于 Gini 计算）
def lorenz_curve(y_true, y_pred):
    """计算 Lorenz 曲线"""
    sorted_indices = np.argsort(y_pred)
    sorted_true = y_true[sorted_indices]

    cum_true = np.cumsum(sorted_true) / np.sum(sorted_true)
    return cum_true
```

```python
%%time
from sklearn.model_selection import GridSearchCV
import xgboost as xgb

# 定义评估函数
def evaluate_model(y_true, y_pred):
    """计算所有模型评估指标"""
    results = {
        "Pearson": calculate_pearson_corr(y_true, y_pred),
        "Spearman": calculate_spearman_corr(y_true, y_pred),
        "Gini": gini(y_true, y_pred),
        "Modified Gini": gini_normalized(y_true, y_pred),
        "KS": calculate_ks(y_true, y_pred),
        "RMSE": calculate_rmse(y_true, y_pred),
        "Hosmer-Lemeshow": calculate_hosmer_lemeshow(y_true, y_pred),
        "Accuracy Index": calculate_accuracy_index(y_true, y_pred)
    }
    return results

# 创建 XGBoost 分类器
xgb_clf = xgb.XGBClassifier(objective='binary:logistic', use_label_encoder=False, eval_metric='auc')

# 定义超参数搜索空间
param_grid = {
    'n_estimators': [100, 300, 500],
    'learning_rate': [0.05, 0.1, 0.3, 0.5, 0.7],
    'max_depth': [3, 6, 9]
}

# 进行超参数搜索
grid_search = GridSearchCV(
    estimator=xgb_clf,
    param_grid=param_grid,
    scoring='roc_auc',
    cv=5,
    n_jobs=-1,
    verbose=2
)

# 训练模型
grid_search.fit(X_train, y_train)

# 初始化存储超参数 + 训练/测试评估指标的 DataFrame
results_list = []

# 遍历所有超参数组合
for params, mean_score, std_score in zip(
    grid_search.cv_results_["params"],
    grid_search.cv_results_["mean_test_score"],
    grid_search.cv_results_["std_test_score"]
):
    # 用当前超参数训练模型（⚠️ 这里不需要重新训练，只进行预测）
    model = xgb.XGBClassifier(**params)
    model.fit(X_train, y_train)  # 直接 fit，不需要再次 grid_search.fit() 

    # **计算训练集指标**
    y_train_pred = model.predict_proba(X_train)[:, 1]
    train_metrics = evaluate_model(y_train, y_train_pred)
    train_metrics = {f"train_{k}": v for k, v in train_metrics.items()}  # 训练集指标加 "train_" 前缀

    # **计算测试集指标**
    y_test_pred = model.predict_proba(X_test)[:, 1]
    test_metrics = evaluate_model(y_test, y_test_pred)
    test_metrics = {f"test_{k}": v for k, v in test_metrics.items()}  # 测试集指标加 "test_" 前缀

    # **存入 DataFrame**
    results_list.append({**params, "ROC_AUC": mean_score, "ROC_AUC_STD": std_score, **train_metrics, **test_metrics})

# **转换为 DataFrame 并输出**
results_df = pd.DataFrame(results_list)
```

```python
from autogluon.tabular import TabularPredictor
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
target = "depvar"
train_data, val_data = train_test_split(df1, test_size=0.2, random_state=42)
```

```python
%%time
predictor = TabularPredictor(
    label=target,
    problem_type="binary",  # 任务类型
    eval_metric="roc_auc",  # 选择评估指标
    path="./AutoGluonModels/",  # 指定存储路径
).fit(train_data, presets="best_quality")

# 分别在训练集和验证集上评估模型性能
train_perf = predictor.evaluate(train_data)
val_perf = predictor.evaluate(val_data)

# 输出性能
print("📊 模型在 **训练集** 上的性能:", train_perf)
print("📊 模型在 **验证集** 上的性能:", val_perf)
```
